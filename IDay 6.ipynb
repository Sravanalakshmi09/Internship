### Data Preprocessing

1.  **Rename columns**: Rename `v1` to `label` and `v2` to `text` for better readability.
2.  **Handle missing values**: Check and remove any rows with missing data.
3.  **Encode labels**: Convert categorical labels ('ham', 'spam') into numerical format (0, 1).

```python
# -------------------------------
# Data Preprocessing
# -------------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Remove rows with missing values
df = df.dropna()

# Display first 5 rows
print("\nFirst 5 Rows:")
print(df.head())

# Display dataset information
print("\nDataset Information:")
print(df.info())

# Display column names
print("\nColumns:")
print(df.columns.tolist())

# Convert categorical columns to numeric
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col].astype(str))

print("\nPreprocessing Completed Successfully!")
print(df.head())
```

### Split Data and Feature Extraction

1.  **Split the data**: Divide the dataset into training and testing sets.
2.  **TF-IDF Vectorization**: Convert the text data into numerical feature vectors using TF-IDF (Term Frequency-Inverse Document Frequency).

```python
# Define features (X) and target (y)
# 'Anomaly' is the target variable, others are features

if 'Anomaly' in df.columns:
    X = df.drop('Anomaly', axis=1)
    y = df['Anomaly']

    # Split the data into training and testing sets
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    print("Data split into training and testing sets successfully!")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape: {X_test.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"y_test shape: {y_test.shape}")
else:
    print("Error: The 'Anomaly' column was not found in the DataFrame.")
    print("Current columns in df:", df.columns.tolist())
    print("Please ensure 'Anomaly' is present in your DataFrame before splitting.")
```

### Train Logistic Regression Model

Initialize and train a Logistic Regression model on the TF-IDF transformed training data.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Initialize the model
model = RandomForestClassifier(random_state=42)

# 2. Train the model using the training sets created in the previous cell
model.fit(X_train, y_train)

# 3. Make predictions on the correct test set (X_test, not X_test_tfidf)
y_pred = model.predict(X_test)

# 4. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(report)
```
